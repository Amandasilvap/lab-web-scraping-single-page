```python
#initialize empty lists
title = []
dir_stars = []
year = []


num_iter = len(soup.select("td.titleColumn a"))

for i in range(num_iter):
    title.append(soup.select("td.titleColumn a")[i].get_text())
    dir_stars.append(soup.select("td.titleColumn a")[i]["title"])
    year.append(soup.select("td.titleColumn span.secondaryInfo")[i].get_text())
    
print(title)
print(dir_stars)
print(year)

movies = pd.DataFrame({"title":title, "dir_stars":dir_stars,"year":year})
```

```python
director = []
star_1 = []
star_2 = []

for movie in dir_stars:
    crew= movie.split(",")
    director.append(crew[0].replace(" (dir.)",""))
    star_1.append(crew[1])
    star_2.append(crew[2])
```

```python
year_c = []

for iten_year in year:
    iten_year = iten_year.replace('(', '').replace(')','')
    year_c.append(iten_year)
```

```python
movies = pd.DataFrame({"title":title,
                       "director":director,
                       "star_1":star_1,
                       "star_2":star_2,
                       "Year":year_c
                      })
movies.head()
```

```python
## Assembling the script to send and store multiple requests
```

```python
iterations = range(1,5)
```

```python
pages = []

for i in iterations:
    # assemble the url:
    start_at= str(i)
    url = "https://www.imdb.com/search/title/?title_type=feature&release_date=1990-01-01,1992-12-31&user_rating=7.5,&start=" + start_at + "&ref_=adv_nxt&lang=eng"

    # download html with a get request:
    response = requests.get(url)
    #response = requests.get(url, headers = {"Accept-Language": "en-US"})

    # monitor the process by printing the status code
    print("Status code: " + str(response.status_code))

    # store response into "pages" list
    pages.append(response)

    # respectful nap:
    wait_time = randint(1,4)
    print("I will sleep for " + str(wait_time) + " second/s.")
    sleep(wait_time)
```

```python
pages_parsed = []
titles = []
synopsis = []

for i in range(len(pages)):
    # parse all pages
    pages_parsed.append(BeautifulSoup(pages[i].content, "html.parser"))
    # select only the info about the movies
    movies_html = pages_parsed[i].select("div.lister-item-content")
    # for movie, store titles and reviews into lists
    for j in range(len(movies_html)):
        titles.append(movies_html[j].select("h3 > a")[0].get_text())
        synopsis.append(movies_html[j].select("p:nth-of-type(2)")[0].get_text().strip())

# Checking our output:

print(len(titles)) # output: 631
print(len(synopsis))  # output: 631
```

```python
# 1. import libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd

# 2. find url and store it in a variable
url = "https://en.wikipedia.org/wiki/List_of_presidents_of_the_United_States"


# 3. download html with a get request
response = requests.get(url)
response.status_code # 200 status code means OK!

# 4.1. parse html (create the 'soup')
soup = BeautifulSoup(response.content, "html.parser")

# 4.2. check that the html code looks like it should
soup
```

```python
soup.select("#mw-content-text > div.mw-parser-output > table.wikitable > tbody > tr:nth-of-type(2) > td:nth-of-type(3) > b > a")
```

```python
# this solution is not very elegant, but works. The CSS selector we copied had an "nth-child" that we could iterate to find presidents, but some elements were empty, so we concatenate each new element with "+" instead of appending as usual:
presidents = []

for i in range(100):
    presidents = presidents + soup.select("#mw-content-text > div.mw-parser-output > table.wikitable > tbody > tr:nth-of-type(" + str(i+1) + ") > td:nth-of-type(3) > b > a")

# check the output:
presidents
```

```python
presidents[0]["href"]
```

```python
# send request
url = "https://en.wikipedia.org/" + presidents[0]["href"]
response = requests.get(url)
print(response.status_code)

# parse & store html
soup = BeautifulSoup(response.content, "html.parser")
soup.select("table.infobox.vcard")
```

```python
import random
from time import sleep
```

```python
# 2. find url and store it in a variable

presi_soups = []

for presi in presidents:
    # send request
    url = "https://en.wikipedia.org/" + presi["href"]
    response = requests.get(url)
    print(presi.get_text(), response.status_code)

    # parse & store html
    soup = BeautifulSoup(response.content, "html.parser")
    presi_soups.append(soup.select("table.infobox.vcard"))

    # respectful nap:
    wait_time = random.randint(1,4)
    print("I will sleep for " + str(wait_time) + " second/s.")
    sleep(wait_time)
```

```python
president_name = []
birthdays = []
parties = []
children = []

for presi in presi_soups:
    birthdays.append(presi[0].select("span.bday")[0].get_text())
    parties.append(presi[0].find("th", string = "Political party").parent.select('a')[0].get_text())
    president_name.append(presi[0].select("div.fn")[0].get_text())
    try:
        children.append(len(presi[0].find("th", string="Children").parent.find_all("li")))
    except:
        children.append("NA")

presis_df = pd.DataFrame({"name":president_name,
                          "birthday":birthdays,
                          "party": parties,
                          "children": children})
```
